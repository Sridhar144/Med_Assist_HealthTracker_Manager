```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('ntrain_data.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df['prognosis'].unique()
```

```python
X = df.drop('prognosis', axis=1)
y = df['prognosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
X_train.shape
```

```python
accuracy = {}
best_param = {}

def model_testing(model_name, model, param_grid_):
    grid = GridSearchCV(model, param_grid=param_grid_)
    grid.fit(X_train, y_train)
    y_pred = grid.predict(X_test)
    
    accuracy[model_name] = accuracy_score(y_test, y_pred)
#     con_mat = confusion_matrix(y_test, y_pred)
#     conf_mat[model_name] = con_mat
    
    best_param[model_name] = grid.best_params_
```

```python
# Logistic Regression
log_model = LogisticRegression(solver='saga',multi_class="ovr",max_iter=1000)
param_grid_log = {'C': np.logspace(0, 4, 5), 'penalty': ['l1', 'l2']}
model_testing('Logistic_Regression', log_model, param_grid_log)
```

```python
svm_model = SVC()
param_grid_svm = {'kernel': ['linear', 'rbf', 'poly'],
             'C': [0.01, 10, 100]}
model_testing('SVM', svm_model, param_grid_svm)
```

```python
cols = ['itching', 'skin_rash', 'continuous_sneezing', 'shivering', 'chills', 'joint_pain', 'cough', 'vomiting', 'headache', 'chest_pain', 'breathlessness', 'nausea', 'diarrhoea', 'muscle_pain', 'sweating']
```

```python
def get_disease_info(predicted_disease):
    # Read the text file
    with open('Diseases_Info.txt', 'r') as file:
        data = file.read()

    disease_sections = data.split('\n\n')

    for section in disease_sections:
        if predicted_disease.lower() in section.lower():
            return section

    return f"Information for {predicted_disease} not found."



my_list = [0]*132

itch = int(input("1 or 0: Itch "))
my_list[d_in['itching']] = itch

skin_rash = int(input("1 or 0: skin rash "))
my_list[d_in['skin_rash']] = itch

con_sne = int(input("1 or 0: continuous_sneezing "))
my_list[d_in['continuous_sneezing']] = con_sne

shiver = int(input("1 or 0: shivering "))
my_list[d_in['shivering']] = shiver

chill = int(input("1 or 0: chills "))
my_list[d_in['chills']] = chill

j_p = int(input("1 or 0: joint_pain "))
my_list[d_in['joint_pain']] = j_p

vom = int(input("1 or 0: vomiting "))
my_list[d_in['vomiting']] = vom

cou = int(input("1 or 0: cough "))
my_list[d_in['cough']] = cou

head = int(input("1 or 0: headache "))
my_list[d_in['headache']] = head

c_p = int(input("1 or 0: chest_pain "))
my_list[d_in['chest_pain']] = c_p

b_l = int(input("1 or 0: chest_pain "))
my_list[d_in['breathlessness']] = b_l

nau = int(input("1 or 0: nausea "))
my_list[d_in['nausea']] = nau

dia = int(input("1 or 0: diarrhoea "))
my_list[d_in['diarrhoea']] = dia

mp = int(input("1 or 0: muscle_pain "))
my_list[d_in['muscle_pain']] = mp

sweat = int(input("1 or 0: muscle_pain "))
my_list[d_in['sweating']] = sweat


outcome = log_model.predict([my_list])[0]
print(get_disease_info(outcome))
```
